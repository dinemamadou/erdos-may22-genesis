```python
## For data handling
import pandas as pd
import numpy as np

## For plotting
import matplotlib.pyplot as plt
from seaborn import set_style

## This sets the plot style
## to have a grid on a white background
set_style("whitegrid")
```

```python
df_11 = pd.read_csv('Models_training_data/model_11_training_log.csv')
```

```python

fig, axes = plt.subplots(1, 2, figsize = (14, 4))
df = df_11
mod = 11
    
epochs = len(df)
    
#Plot the accuracy

#axes[0].scatter(range(1,epochs+1), df.accuracy.values*100, label = "Training Accuracy")
#axes[0].scatter(range(1,epochs+1), df.val_accuracy.values*100, label = "Validation Set Accuracy")

axes[0].scatter(range(1,epochs+1), df.accuracy.values*100, label = "Training set")
axes[0].scatter(range(1,epochs+1), df.val_accuracy.values*100, label = "Validation set")

axes[0].set_xlabel("Epoch", fontsize=18)
axes[0].set_ylabel("Accuracy (%)", fontsize=18)
axes[0].set_ybound(0, 100)
    
axes[0].tick_params(labelsize=16)
    

axes[0].legend(fontsize=18)
    
#axes[0].set_title('Final Model Accuracy', fontsize=18)
axes[0].set_title('Accuracy', fontsize=18)



#Plot the loss
    
#axes[1].scatter(range(1,epochs+1), df.loss.values, label = "Training Loss")
#axes[1].scatter(range(1,epochs+1), df.val_loss.values, label = "Validation Loss")

axes[1].scatter(range(1,epochs+1), df.loss.values, label = "Training set")
axes[1].scatter(range(1,epochs+1), df.val_loss.values, label = "Validation set")

axes[1].set_xlabel("Epoch", fontsize=18)
axes[1].set_ylabel("Loss", fontsize=18)
axes[1].set_ybound(0, 3)

axes[1].tick_params(labelsize=16)
    

axes[1].legend(fontsize=18)
    
#axes[1].set_title('Final Model Loss', fontsize=18)
axes[1].set_title('Loss', fontsize=18)



plt.rcParams["axes.linewidth"] = 1.5
plt.rcParams["axes.edgecolor"] = "black"


plt.show()
    
    
    
    
    

```
