```python
from pyAudioAnalysis import ShortTermFeatures as aF
from pyAudioAnalysis import audioBasicIO as aIO 
from pyAudioAnalysis import MidTermFeatures as mF
import numpy as np 
import plotly.graph_objs as go 
import plotly
import IPython

import pandas as pd

import os, shutil
import time

import matplotlib.pyplot as plt

import random
```

```python
# Get list of all file names in AudioWav
files = os.listdir('Data_temp')
files.sort()
```

```python
#Extract the averaged mid_features for all WAV files in the directory AudioWav. 
# Arguments: filename (str), 
#            mid_window (s), 
#            mid_step (s), 
#            short_window (s), 
#            short_step (s), 
#            compute_beat (Bool)
# Returns: features, file_list, mid_feature_names 


features1, file_list1, mid_feature_names1 = mF.directory_feature_extraction('Data_temp', mid_window=0.5, mid_step = 0.1, 
                                                 short_window = .05, short_step = .025,
                                                  compute_beat=False)
```

```python
features1[0]
```

```python
# Create a data frame. Among the columns should be: name of file, mid feature values, classification, actor id


# Create a list of file names, with just actorID, emotion, and intensity:
nice_names = []
for name in files:
    stripped_name = name.lstrip('Data_temp')
    stripped_name = stripped_name.strip('.wav')
    nice_names.append(stripped_name)

```

```python
nice_names
```

```python
# Create new columns for Actor ID, emotion, and sentence

actor_ids = []
emotion = []
sentence = []


for name in nice_names:
    list = name.split('_')
    actor_ids.append(list[0])
    sentence.append(list[1])
    emotion.append(list[2])
```

```python
# Create a dict with keys the column names of the data frame
    
dict = {}
dict['FileID'] = nice_names
dict['actorID'] = actor_ids
dict['Emotion'] = emotion
dict['SentenceID'] = sentence

    
# add columns for the mid features to the data frame. 
     


for i in range(len(mid_feature_names1)):
    dict[mid_feature_names1[i]] = features1[:, i]
```
