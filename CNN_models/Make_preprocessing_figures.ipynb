```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from matplotlib.pyplot import imread, imshow, subplots, show


```

```python
def plot(data_generator):
    """
    Plots 4 images generated by an object of the ImageDataGenerator class.
    """
    data_generator.fit(images)
    image_iterator = data_generator.flow(images)
    
    # Plot the images given by the iterator
    fig, rows = subplots(nrows=1, ncols=4, figsize=(18,18))
    for row in rows:
        row.imshow(image_iterator.next()[0])
        row.axis('off')
    show()
    




```

```python
image = imread('1010_TSI_ANG_XX.png')

# Creating a dataset which contains just one image.
images = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))


#Show the original image:

imshow(images[0])
show()
```

```python
# Make a generator to see the effect of height_shift:


data_generator = ImageDataGenerator(height_shift_range = 0.2,
                                   fill_mode = 'wrap')
```

```python

#Display different "height_shift" images. Why are they horizontally shifted, instead of vertically?

plot(data_generator)
```

```python
# Generator to see the effect of "width_shift": 


data_generator_2 = ImageDataGenerator(width_shift_range = 0.2,
                                     fill_mode = 'wrap')
```

```python
# Display different "width_shift" images. Why are they vertically shifted, instead of horizontally? 


plot(data_generator_2)
```
