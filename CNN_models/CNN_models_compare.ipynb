```python
## For data handling
import pandas as pd
import numpy as np

## For plotting
import matplotlib.pyplot as plt
from seaborn import set_style

## This sets the plot style
## to have a grid on a white background
set_style("darkgrid")
```

```python
df_1 = pd.read_csv('model_training_data/model_1_training_log.csv')
df_2 = pd.read_csv('model_training_data/model_2_training_log.csv')
df_3 = pd.read_csv('model_training_data/model_3_training_log.csv')
df_4 = pd.read_csv('model_training_data/model_4_training_log.csv')
df_5 = pd.read_csv('model_training_data/model_5_training_log.csv')
df_6 = pd.read_csv('model_training_data/model_6_training_log.csv')
df_8 = pd.read_csv('model_training_data/model_8_training_log.csv')
df_9 = pd.read_csv('model_training_data/model_9_training_log.csv')
df_11 = pd.read_csv('model_training_data/model_11_training_log.csv')
df_12 = pd.read_csv('model_training_data/model_12_training_log.csv')




df_1.head()
```

Model 1: “normal” sized layers, added dropout layer (rate = 0.5), preprocessing: width shift, range 0.2, with fill = ‘wrap’. 

```python
plt.figure(figsize = (10,6))

epochs = len(df_1)

plt.scatter(range(1,epochs+1), df_1.accuracy.values, label = "Training Accuracy")
plt.scatter(range(1,epochs+1), df_1.val_accuracy.values, label = "Validation Set Accuracy")

plt.xlabel("Epoch", fontsize=18)
plt.ylabel("Accuracy", fontsize=18)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.legend(fontsize=18)
plt.title('Model 1 Accuracy', fontsize=18)

plt.show()


plt.figure(figsize = (10,6))

plt.scatter(range(1,epochs+1), df_1.loss.values, label = "Training Loss")
plt.scatter(range(1,epochs+1), df_1.val_loss.values, label = "Validation Set Loss")

plt.xlabel("Epoch", fontsize=18)
plt.ylabel("Loss Function Value", fontsize=18)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.legend(fontsize=18)
plt.title('Model 1 Loss', fontsize=18)

plt.show()
```

```python
dfs = [df_1, df_2, df_3, df_4, df_5, df_6, df_8, df_9, df_11, df_12]

mod_nos = [1, 2, 3, 4, 5, 6, 8, 9, 11, 12]

num_mods = len(dfs)

for i in range(num_mods):
    fig, axes = plt.subplots(1, 2, figsize = (14, 4))
    df = dfs[i]
    mod = mod_nos[i]
    
    epochs = len(df)
    
    #Plot the accuracy

    axes[0].scatter(range(1,epochs+1), df.accuracy.values, label = "Training Accuracy")
    axes[0].scatter(range(1,epochs+1), df.val_accuracy.values, label = "Validation Set Accuracy")

    axes[0].set_xlabel("Epoch", fontsize=18)
    axes[0].set_ylabel("Accuracy", fontsize=18)
    axes[0].set_ybound(0, 0.8)
    
    axes[0].tick_params(labelsize=16)
    

    axes[0].legend(fontsize=18)
    
    axes[0].set_title('Model ' +str(mod) +  ' Accuracy', fontsize=18)
    
    #Plot the loss
    
    axes[1].scatter(range(1,epochs+1), df.loss.values, label = "Training Loss")
    axes[1].scatter(range(1,epochs+1), df.val_loss.values, label = "Validation Loss")

    axes[1].set_xlabel("Epoch", fontsize=18)
    axes[1].set_ylabel("Loss", fontsize=18)
    axes[1].set_ybound(0, 3)

    axes[1].tick_params(labelsize=16)
    

    axes[1].legend(fontsize=18)
    
    axes[1].set_title('Model ' +str(mod) +  ' Loss', fontsize=18)

    plt.show()
    
    
    
    print()
    


```
